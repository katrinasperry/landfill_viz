```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Load data

```python
df = pd.read_excel('MATERIALS_MUNICIPAL_WASTE_STREAM_1960_TO_2014.XLSX')
```

```python
df
```

```python
# drop last 3 rows
df = df[0:17]
```

```python
df
```

```python
# reset index
df.set_index('Materials',inplace=True)
```

```python
df
```

```python
df_t = df.transpose()
```

```python
df_t
```

# Start plotting

```python
df_t.plot(figsize=(20,10))
```

It's hard to see what's going on at the bottom of the plot since the quantities are so different. Let's make separate dataframes of products, other wastes, and totals so we can see what's happening better.

```python
# dataframe of product categories, drop Products - Total Materials column
products = df_t[['Products - Paper and Paperboard', 'Products - Glass', 'Products - Metals - Ferrous',
                'Products - Metals - Aluminum', 'Products - Metals - OtherNonferrous', 'Products - Metals - Total',
                'Products - Plastics', 'Products - Rubber and Leather', 'Products - Textiles', 'Products - Wood',
                'Products - Other']]
```

```python
products.plot(figsize=(20,10))
```

It looks as though glass is the only category that has decreased over time, which makes since from what I've observed. It could be that more and more packaging is plastic rather than glass, and it could also be due to more glass being recycled than put in landfills. It would be interesting to untangle this by looking at recycling data.

```python
# dataframe of other wastes, drop Total MSW Generated - Weight column
wastes = df_t[['Other Wastes - Food Waste', 'Other Wastes - Yard Trimmings', 
              'Other Wastes - Miscellaneous Inorganic Wastes']]
```

```python
wastes.plot(figsize=(20,10))
```

```python
# dataframe of both total columns
totals = df_t[['Products - Total Materials', 'Total MSW Generated - Weight']]
```

```python
totals.plot(figsize=(20,10))
```

```python
sns.pairplot(df_t)
```

Yikes, this is a huge distplot, but it's super interesting! Without inlarging it, you can clearly that most of the plots are upward sloping diagonal(ish) lines. There is just one column and one row that defy this pattern: glass. Nearly every distplot with glass starts in the lower left corner and then bends back toward the x or y-axis. This indicates that glass is decreases in relation to each other category, which confirms the overall decrease in glass in landfills that we saw above.
